```python
# Import necessary libraries
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Configure Matplotlib and Seaborn
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Set the directory path where all 34 files are located
directory_path = "C:/Users/adnan/Downloads/PROMISE-repo/promise/PROMISE"  # Update this to your directory path

# Define thresholds for assigning methodologies
complexity_threshold = 15  # Weighted Methods per Class (WMC) threshold
size_threshold = 1000      # Lines of Code (LOC) threshold


```

```python
# Loop through each file in the directory
for file_name in os.listdir(directory_path):
    if file_name.endswith(".csv"):
        print(f"Processing file: {file_name}")
        
        # Load the data
        file_path = os.path.join(directory_path, file_name)
        data = pd.read_csv(file_path)
        
        # Display initial few rows of the dataset
        display(data.head())

        # Assign methodologies
        data['methodology'] = 'Agile'  # Default to Agile
        data.loc[(data['wmc'] > complexity_threshold) | (data['loc'] > size_threshold), 'methodology'] = 'Waterfall'

        # Save the updated data to a new file
        output_file_path = os.path.join(directory_path, f"updated_{file_name}")
        data.to_csv(output_file_path, index=False)
        print(f"Updated file saved as: {output_file_path}\n")
```

```python
# Collect all updated files for combined analysis
all_data = []

# Load all updated files into a list
for file_name in os.listdir(directory_path):
    if file_name.startswith("updated_") and file_name.endswith(".csv"):
        file_path = os.path.join(directory_path, file_name)
        data = pd.read_csv(file_path)
        all_data.append(data)

# Combine all data into a single DataFrame
combined_data = pd.concat(all_data, ignore_index=True)
```

```python
# Descriptive statistics of the combined data
print("\nDescriptive Statistics for Combined Data:")
display(combined_data.describe())
```

```python
# Correlation analysis
print("\nCorrelation Matrix:")

# Select only numeric columns for correlation calculation
numeric_data = combined_data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Plot the correlation matrix as a heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Prepare features (X) and target (y)
X = combined_data[['wmc', 'loc']]  # Features
y = combined_data['methodology'].apply(lambda x: 1 if x == 'Waterfall' else 0)  # Target: 1 for Waterfall, 0 for Agile

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a logistic regression model
log_model = LogisticRegression()
log_model.fit(X_train, y_train)

# Make predictions
y_pred = log_model.predict(X_test)

# Evaluate the model
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap="Blues")
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
import pandas as pd
import glob

# Path to the folder containing the 34 CSV files
file_path = 'C:/Users/adnan/Downloads/PROMISE-repo/promise/PROMISE/*.csv'  # Replace 'path_to_your_folder' with the actual path

# Load and concatenate all CSV files into one DataFrame
all_files = glob.glob(file_path)
combined_data = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)

# Preview the combined data
print(combined_data.head())
```

```python
# Correlation between complexity metrics and defect rate
complexity_defect_corr = combined_data[['wmc', 'cbo', 'rfc', 'bug']].corr()
print("Complexity vs Defect Rate Correlation:\n", complexity_defect_corr)
```

```python
# Calculate defect density
combined_data['defect_density'] = combined_data['bug'] / combined_data['loc']
defect_density = combined_data.groupby('methodology')['defect_density'].mean()
print("Defect Density by Methodology:\n", defect_density)
```

```python
# Average complexity and loc by methodology
complexity_productivity = combined_data.groupby('methodology').agg({
    'wmc': 'mean', 
    'cbo': 'mean', 
    'loc': 'mean'
})
print("Complexity and LOC by Methodology:\n", complexity_productivity)
```

```python
# Cohesion and coupling metrics by methodology
quality_metrics = combined_data.groupby('methodology').agg({
    'cam': 'mean', 
    'cbo': 'mean'
})
print("Quality Metrics (CAM and CBO) by Methodology:\n", quality_metrics)
```

```python
# Categorize complexity and analyze defect rates
combined_data['complexity_level'] = pd.cut(combined_data['wmc'], bins=[0, 10, 20, 30, 40], labels=['Low', 'Medium', 'High', 'Very High'])
defect_by_complexity = combined_data.groupby(['methodology', 'complexity_level'])['bug'].mean()
print("Defect Rate by Complexity Level and Methodology:\n", defect_by_complexity)
```

```python
# Correlation between CAM (cohesion) and defect rate
cohesion_defect_corr = combined_data[['cam', 'bug']].corr()
print("Cohesion vs Defect Rate Correlation:\n", cohesion_defect_corr)
```

```python
# Correlation between WMC (methods per class) and defect rate
method_defect_corr = combined_data[['wmc', 'bug']].corr()
print("Method Count vs Defect Rate Correlation:\n", method_defect_corr)
```

```python
# Average LOC per class by methodology
avg_loc_per_class = combined_data.groupby('methodology')['loc'].mean()
print("Average LOC per Class by Methodology:\n", avg_loc_per_class)
```
